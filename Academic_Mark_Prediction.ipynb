```python
# Convert 'object' columns with numerical entries to numeric
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        except Exception as e:
            print(f"Error in column {col}: {e}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
def score_range(x):
    if x == 45:
        return '190-200'
    elif x >= 40:
        return '180-189'
    elif x >= 35:
        return '170-179'
    elif x >= 30:
        return '160-169'
    elif x >= 25:
        return '150-159'
    else:
        return 'Below 150'

# Apply score ranges for each subject
df['ETW_range'] = df['ETW'].apply(score_range)
df['MTW_range'] = df['MTW'].apply(score_range)
df['HTW_range'] = df['HTW'].apply(score_range)
df['MtTW_range'] = df['MtTW'].apply(score_range)
df['STW_range'] = df['STW'].apply(score_range)
df['CSTW_range'] = df['CSTW'].apply(score_range)
df['SSTW_range'] = df['SSTW'].apply(score_range)
```

```python
# Combine all ranges for prediction (multi-targets)
subject_targets = ['ETW_range', 'MTW_range', 'HTW_range', 'MtTW_range', 'STW_range', 'CSTW_range', 'SSTW_range']
```

```python
# Encode target variables
le = LabelEncoder()
for target in subject_targets:
    df[target] = le.fit_transform(df[target])

# Features (X) - drop non-numeric IDs and targets
targets = df[subject_targets]
X = df.drop(columns=['Student ID'] + subject_targets)
```

```python
# Split the Dataset
X_train, X_test, y_train, y_test = train_test_split(X, targets, test_size=0.2, random_state=42)
```

```python
# Train Random Forest Classifier
rf_models = {}
rf_predictions = {}
accuracy_scores = {}

for subject in subject_targets:
    print(f"Training model for {subject}...")
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_train, y_train[subject])
    y_pred = rf.predict(X_test)

    rf_models[subject] = rf
    rf_predictions[subject] = y_pred
    accuracy_scores[subject] = accuracy_score(y_test[subject], y_pred)

    print(f"Accuracy for {subject}: {accuracy_scores[subject]:.2f}")
    print(classification_report(y_test[subject], y_pred, target_names=le.classes_))
```

```python
# Summarize Results
print("\nModel Accuracy Summary:")
for subject, acc in accuracy_scores.items():
    print(f"{subject}: {acc:.2f}")
```
